```python
import numpy as np
import xarray as xr

from tqdm.notebook import trange
```

```python
ds = xr.open_dataset("F:/GLEAM/v42a/Ep/Ep_1980_GLEAM_v4.2a_MO.nc")  # 替换为你的 nc 文件路径
AE_GLEAM = ds["Ep"]
ds.close()

for year in trange(1981, 2023):
    ds = xr.open_dataset(f"F:/GLEAM/v42a/Ep/Ep_{str(year)}_GLEAM_v4.2a_MO.nc")  # 替换为你的 nc 文件路径
    temp_AE_GLEAM = ds["Ep"]
    AE_GLEAM = xr.concat([AE_GLEAM, temp_AE_GLEAM], dim='time')

AE_GLEAM.to_netcdf("F:/GLEAM/v42a/Ep/Ep_GLEAM_merged.nc")
```

```python
ds = xr.open_dataset("F:/GLEAM/v42a/E/05/E_1980_05.nc")  # 替换为你的 nc 文件路径
AE_GLEAM = ds["E"]
ds.close()

for year in trange(1981, 2023):
    ds = xr.open_dataset(f"F:/GLEAM/v42a/E/05/E_{str(year)}_05.nc")  # 替换为你的 nc 文件路径
    temp_AE_GLEAM = ds["E"]
    AE_GLEAM = xr.concat([AE_GLEAM, temp_AE_GLEAM], dim='time')

AE_GLEAM.to_netcdf("F:/GLEAM/v42a/E/05/E_GLEAM_merged.nc")
```

```python
new_AE_GLEAM = np.rot90(AE_GLEAM.values, axes=(1, 2))  # 转换为 numpy 数组并旋转纬度和经度轴
```

```python
np.save("../../Data/GLEAM/E_GLEAM_1981_2023.npy", new_AE_GLEAM)
```
